```python
import numpy as np
import tools
from matplotlib import pyplot as plt

card_to_track = 3
number_of_cards = 52
deck = np.arange(number_of_cards)

iterations = 50_000
number_of_shuffles = 3

positions = []
positionsTally = np.zeros(number_of_cards)

for i in range(iterations):
    newOrder = deck
    
    for s in range(number_of_shuffles):
        newOrder = list(tools.gsr_riffle_shuffle(newOrder))
        
    newPosition = newOrder.index(card_to_track)
    positions.append(newPosition)
    positionsTally[newPosition] = positionsTally[newPosition] + 1
    
proportions = positionsTally / iterations

# TODO use positionsTally
plt.hist(positions, bins=number_of_cards)
plt.show()

# TODO reverse process - probability card came from position?
```
