```python
import numpy as np
import tools
import random
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
number_of_cards = 52

# this can be increased
# 1_000_000 works well for large model
# set to a lower number for faster development
number_of_rows = 100_000 
test_train_split = 0.8
```

```python
deck = np.arange(number_of_cards, dtype=int)

# generate training data
X = np.zeros((number_of_rows, len(deck)), dtype=np.uint8)
y = np.zeros((number_of_rows, 1), dtype=np.uint8)

for i in range(number_of_rows):
    _class = random.randint(0, 1) # 1 is riffle, 0 is overhand
    
    random_cut_point = random.randint(0, len(deck)-1)
    
    y[i,:] = _class
    if (_class == 1):
        X[i,:] = tools.cut(tools.gsr_riffle_shuffle(deck), random_cut_point)  
    else:
        X[i,:] = tools.cut(tools.overhand_shuffle(deck), random_cut_point)
```

```python
# show some examples
no_of_examples = 8
examples = np.zeros((no_of_examples, len(deck)))
example_classes = np.zeros((no_of_examples,), dtype=np.uint8)
for i in range(no_of_examples):
    random_index = random.randint(0, X.shape[0] - 1)
    examples[i] = X[random_index,:]
    example_classes[i] = np.uint8(y[random_index][0] > 0.5)
    
print(example_classes)
tools.visualise_multiple_decks(examples, display)

X_train, X_test = tools.split_test_train(X, test_train_split)
y_train, y_test = tools.split_test_train(y, test_train_split)
```

```python
model = Sequential([
    Dense(25, activation="sigmoid"),
    Dense(15, activation="sigmoid"),
    Dense(1, activation="sigmoid"),
])

model.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.006),
    metrics = 'accuracy'
)
```

```python
model.fit(X_train, y_train, epochs=3)
loss, accuracy = model.evaluate(X_test, y_test)
```

```python
# show some examples
no_of__test_examples = 10
test_examples = np.zeros((no_of__test_examples, len(deck)))
text_example_classes = np.zeros((no_of__test_examples,), dtype=np.uint8)

for i in range(no_of__test_examples):
    random_cut_point = random.randint(0, len(deck)-1)
    riffled = tools.cut(tools.gsr_riffle_shuffle(deck), random_cut_point)
    prediction = model.predict(riffled.reshape(1,-1), verbose=0) > 0.5 # expect True
    
    test_examples[i] = riffled
    text_example_classes[i] = int(prediction)

print(text_example_classes)
tools.visualise_multiple_decks(test_examples, display)

for i in range(no_of__test_examples):
    random_cut_point = random.randint(0, len(deck)-1)
    overhand_shuffled = tools.cut(tools.overhand_shuffle(deck), random_cut_point)
    prediction = model.predict(overhand_shuffled.reshape(1,-1), verbose=0) > 0.5 # expect False
    
    test_examples[i] = overhand_shuffled
    text_example_classes[i] = int(prediction)
    
print(text_example_classes)
tools.visualise_multiple_decks(test_examples, display)
```

```python
def to_sparse_matrix(deck):
    # todo bool
    x = np.zeros((len(deck), len(deck)), dtype=bool)
    for i in range(len(deck)):
        x[deck[i], i] = True
    
    return x

X_train_large = np.array(list(map(lambda d: np.array(to_sparse_matrix(d).ravel()), X_train)), dtype=bool)
X_test_large = np.array(list(map(lambda d: np.array(to_sparse_matrix(d).ravel()), X_test)), dtype=bool)

large_model = Sequential([
    Dense(1000, activation="sigmoid"),
    Dense(500, activation="sigmoid"),
    Dense(1, activation="sigmoid"),
])

large_model.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.006),
    metrics = 'accuracy'
)

large_model.fit(X_train_large, y_train, epochs=1)

large_model.evaluate(X_test_large, y_test)

# save the model!
large_model.save("models/first-model.keras")
```

```python
# use the saved model
restored_model = tf.keras.models.load_model("models/first-model.keras")
restored_model.summary()


loss, acc = restored_model.evaluate(X_test_large, y_test)
display(f"Restored model accuracy is {acc*100:.2f}%")
```
